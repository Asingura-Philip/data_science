# Task 3 - GroupBy and Pivot Tables

Group the parameters "GPU", "CPU_core" and "Price" to make a pivot table and visualize this connection using the pcolor plot.


```python
# Write your code below and press Shift+Enter to execute
# Create the group
df_gptest = df[['GPU','CPU_core','Price']]
Room2_grouped_test = df_gptest.groupby(['GPU','CPU_core','Price'],as_index=False).mean()
Room2_grouped_test
```

```python
# Write your code below and press Shift+Enter to execute


# Creating the Pivot table
grouped_pivot = Room2_grouped_test.pivot_table(index='CPU_core',values='Price',columns='GPU')
grouped_pivot
```

```python
# Write your code below and press Shift+Enter to execute
# Create the Plot
plt.pcolor(grouped_pivot, cmap='RdBu')
plt.colorbar()
plt.show()
```

# Task 4 - Pearson Correlation and p-values

Use the `scipy.stats.pearsonr()` function to evaluate the Pearson Coefficient and the p-values for each parameter tested above. This will help you determine the parameters most likely to have a strong effect on the price of the laptops.


```python
# Write your code below and press Shift+Enter to execute
from scipy import stats


# looking for the pearson coefficient and p valuse between GPU and Price
pearson_coef, p_value = stats.pearsonr(df['GPU'], df['Price'])
print("The Pearson Correlation Coefficient is", pearson_coef, " with a P-value of P =", p_value)
```

```python
# looking for the pearson coefficient and p valuse between CPU_core and Price
pearson_coef, p_value = stats.pearsonr(df['CPU_core'], df['Price'])
print("The Pearson Correlation Coefficient is", pearson_coef, " with a P-value of P =", p_value)
```
